```python
from pyspark.context import SparkContext
from pyspark.sql.session import SparkSession
sc = SparkContext('local')
spark = SparkSession(sc)
```

```python
# from pyspark.sql import SparkSession
# spark = SparkSession.builder.getOrCreate()
```

```python
KSA_GT_path="KSA_GT.csv"
KSA_MT_path="KSA_MT.csv"
UAE_GT_MT_path="UAE_MT_GT.csv"
QTR_GT_MT_path="QTR_MT_GT.csv"
```

```python
KSA_GT = spark.read.options(header='True', delimiter=',').csv(KSA_GT_path)
KSA_MT = spark.read.options(header='True', delimiter=',').csv(KSA_MT_path)
UAE_GT_MT = spark.read.options(
    header='True', delimiter=',').csv(UAE_GT_MT_path)
QTR_GT_MT = spark.read.options(
    header='True', delimiter=',').csv(QTR_GT_MT_path)
```

```python
KSA_GT=KSA_GT.na.drop("all")
KSA_MT=KSA_MT.na.drop("all")
UAE_GT_MT=UAE_GT_MT.na.drop("all")
QTR_GT_MT=QTR_GT_MT.na.drop("all")
```

```python
from pyspark.sql.functions import col
UAE_GT_MT = UAE_GT_MT.withColumn("Region",col("Country"))
QTR_GT_MT = QTR_GT_MT.withColumn("Region",col("Country"))
```

```python
KSA_GT.createOrReplaceTempView('KSA_GT_2')
KSA_MT.createOrReplaceTempView('KSA_MT_2')
UAE_GT_MT.createOrReplaceTempView('UAE_GT_MT_2')
QTR_GT_MT.createOrReplaceTempView('QTR_GT_MT_2')
```

```python
KSA_GT_2 = spark.sql("""select Country
,'Trade Format'
,Year
,Month
,Week
,'Customer Number' as CustomerCodeBannerCode
,'Customer Name'
,'Stock Location (WH or In Store)'
,'WH Location or Store Name'
,Channel as StoreFormat
,Region
from KSA_GT_2""")
```

```python
KSA_GT_2.toPandas()
```

```python
KSA_MT_2 = spark.sql("""select 
Country
,'Trade Format'
,Year
,MONTH as Month
,Week
,'Customer Account / Banner Code' as CustomerCodeBannerCode
,ACCOUNT as CustomerName
,'ACCOUNT LOCATION' as StockLocationWHorInStore
,'STORE NAME' as WHLocationorStoreName
,'Store Format'
,Region

from KSA_MT_2""")
```

```python
KSA_MT_2.toPandas()
```

```python
UAE_GT_MT_2=spark.sql("""select 
Country
,'Trade Format'
,Year
,Month
,Week
,'Customer Code / Banner Code'
,Customer as CustomerName
,'Store or Warehouse' as StockLocationWHorInStore
,'WH Location or Store Name'
,'Store Format'
,Region
from UAE_GT_MT_2""")
```

```python
UAE_GT_MT_2.toPandas()
```

```python
QTR_GT_MT_2 = spark.sql("""select 
Country
,'Trade Format'
,Year
,Month
,Week
,'Customer Code / Banner Code'
,Customer as CustomerName
,'Store or Warehouse' as StockLocationWHorInStore
,'WH Location or Store Name'
,'Store Format'
,Region
from QTR_GT_MT_2""")
```

```python
QTR_GT_MT_2.toPandas()
```

```python
Result = spark.sql(""" select Country
,'Trade Format'
,Year
,Month
,Week
,'Customer Number' as CustomerCodeBannerCode
,'Customer Name'
,'Stock Location (WH or In Store)'
,'WH Location or Store Name'
,Channel as StoreFormat
,Region
from KSA_GT_2

union all
select 
Country
,'Trade Format'
,Year
,MONTH as Month
,Week
,'Customer Account / Banner Code' as CustomerCodeBannerCode
,ACCOUNT as CustomerName
,'ACCOUNT LOCATION' as StockLocationWHorInStore
,'STORE NAME' as WHLocationorStoreName
,'Store Format'
,Region

from KSA_MT_2

union all
select 
Country
,'Trade Format'
,Year
,Month
,Week
,'Customer Code / Banner Code'
,Customer as CustomerName
,'Store or Warehouse' as StockLocationWHorInStore
,'WH Location or Store Name'
,'Store Format'
,Region
from UAE_GT_MT_2

union all
select 
Country
,'Trade Format'
,Year
,Month
,Week
,'Customer Code / Banner Code'
,Customer as CustomerName
,'Store or Warehouse' as StockLocationWHorInStore
,'WH Location or Store Name'
,'Store Format'
,Region
from QTR_GT_MT_2
""")
```

```python
Result.toPandas()
```
